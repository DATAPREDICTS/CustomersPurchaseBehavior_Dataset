# Customer Sentiment Analysis and Feedback Insights

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Data Collection
df=pd.read_csv('purchase behaviou dataset.csv.zip',encoding='unicode_escape')
```

```python
ax=sns.countplot(x='LIFESTAGE',data=df)
sns.set(rc={'figure.figsize':(30,20)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='PREMIUM_CUSTOMER',data=df)
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
lifestage_customer=df.groupby(['LIFESTAGE'],as_index=False)['PREMIUM_CUSTOMER'].count().sort_values(by='PREMIUM_CUSTOMER',ascending=False)
sns.barplot(x='LIFESTAGE',y='PREMIUM_CUSTOMER',data= lifestage_customer)
sns.set(rc={'figure.figsize':(30,25)})
```
